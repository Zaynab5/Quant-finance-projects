# Monte Carlo Simulation for Option Pricing

## Objective
To estimate the price of a European call option using Monte Carlo simulation.

## Methodology
Simulated stock price paths using Geometric Brownian Motion and computed option payoffs.

## Results
Estimated option price using 10,000 simulations and compared with theoretical expectations.

## Conclusion
Monte Carlo simulation provides a flexible numerical approach for pricing derivatives under uncertainty.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100      # Initial stock price
K = 100       # Strike price
T = 1         # Time to maturity (1 year)
r = 0.05      # Risk-free rate
sigma = 0.2   # Volatility
n_simulations = 10000
n_steps = 252

dt = T / n_steps

# Simulate price paths
np.random.seed(42)
Z = np.random.standard_normal((n_steps, n_simulations))
S = np.zeros_like(Z)
S[0] = S0

for t in range(1, n_steps):
    S[t] = S[t-1] * np.exp((r - 0.5 * sigma**2)*dt + sigma*np.sqrt(dt)*Z[t])

# Option payoff (European Call)
payoff = np.maximum(S[-1] - K, 0)

# Discounted price
option_price = np.exp(-r*T) * np.mean(payoff)

print("Estimated Option Price:", option_price)

# Plot few simulated paths
plt.plot(S[:, :10])
plt.title("Simulated Stock Price Paths")
plt.xlabel("Time Steps")
plt.ylabel("Stock Price")
plt.show()
```
